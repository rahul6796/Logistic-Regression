```python
def input_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if(pd.isnull(Age)):
        if(Pclass==1):
            return 37
        if(Pclass==2):
            return 29
        else:
            return 24
    else:
        return Age
```

```python
data['Age']=data[['Age','Pclass']].apply(input_age,axis=1)
```

```python
# now In pclass column we are going to make it one hot encoded vector:
def one_hot_encoded(col_name):
    unique_val=list(np.unique(col_name))
    encoded=np.zeros((len(col_name),len(unique_val)))
    for i,c in enumerate(col_name):
        encoded[i][unique_val.index(c)]=1
    return encoded
    
```
